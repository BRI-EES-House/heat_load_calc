## 共通部分のサンプリング

共通部分は地域と面積の決定が必要になる。
地域については8地域を均等分布で生成する。
面積についてはひとまずガンマ分布を使う。正規分布のほうが良いかもしれない・・。


```python
def sample_common():
    #居室区分ごとに面積をガンマ分布で生成
    a_m   = gamma(2, 30)
    a_o   = gamma(2, 60)
    a_non = gamma(2, 30)
    return {
        'Region'                     : choice(['1', '2', '3', '4', '5', '6', '7', '8']),
        'IsSimplifiedInput'          : True,
        'MainOccupantRoomFloorArea'  : a_m,
        'OtherOccupantRoomFloorArea' : a_o,
        'TotalFloorArea'             : a_m + a_o + a_non
    }
```

## 外壁のサンプリング

外壁は断熱方法によって生成する壁を変える必要があるかが悩みどころ。
ひとまずは、そこは考えずに単純に屋根・天井・垂直壁を生成する。

垂直壁については生成する数を4以上になるように正規分布を調整する。
その方位は8方位から一様分布で生成させる。

すべての壁の面積はガンマ分布、U値は正規分布とした。


```python
def sample_wall(TotalFloorArea):

    #断熱方法
    insulationType = choice(['FloorIns', 'BaseIns'])
    
    #U値
    uRoof = normal(7.7, 2)
    uWall = normal(6.67, 3)
    uFloor = normal(5.27, 1.5)
    uDoor = normal(4.65, 1.2)
    
    #屋根天井、床
    ceiling = { 'Name': 'Ceiling', 'Type': 'Ceiling', 'Structure': 'Wood', 'Direction': 'Top',  'UA': uRoof,  'Area': gamma(2, 50) }
    floor = { 'Name': 'Floor',   'Type': 'Floor',   'Structure': 'Wood', 'Direction': 'Bottom', 'UA': uFloor, 'Area': gamma(2, 45) }
    
    #垂直壁
    nVertWall = math.floor(max(4, normal(4, 2)))
    parts_vertwalls =  [
      { 
        'Name': 'Wall_{}'.format(no),
        'Type': 'Wall',
        'Structure': 'Wood',
        'Direction': choice(['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW']),
        'UA': uWall,
        'Area': gamma(2, 30)
      }
      for no in range(nVertWall)
    ]
    
    #パーツ
    parts = [ceiling] + parts_vertwalls + [floor]
    
    return [ make_wall(TotalFloorArea = TotalFloorArea, **f) for f in parts ]
```

## 窓のサンプリング

窓は外壁とほぼ同じ。

F値を入力するかについて、二項分布で決定する。
η値、F値は正規分布とした。
窓面積は、ガンマ分布とした。


```python
def sample_window( TotalFloorArea, Region ):
    
    UWnd = normal(3.49, 1)
    EtaDC = normal(0.51, 0.2)       
    EtaDH = normal(0.51, 0.2)
    IsInputFValue = binomial(1,0.5) == 1   #二項分布によるF値入力の決定
    FValueC = normal(0.9, 0.2)
    FValueH = normal(0.7, 0.2)
    
    n  = math.floor(max(4, normal(4, 2)))
    parts =  [
      { 
        'Name': 'Window_{}'.format(no),
        'Direction': choice(['N', 'NE', 'E', 'SE', 'S', 'SW', 'W', 'NW']),
        'Area': gamma(2, 20)
      }
      for no in range(n)
    ]
    
    # 簡易外皮入力モデルにおける日除けの寸法を設定
    y1 = 0    # 窓上端と外部日除けの距離(m)   
    y2 = 1.1  # 開口部高さ(m), 腰高窓を想定
    
    return [ make_window(
                TotalFloorArea = TotalFloorArea,
                EtaDC = EtaDC,
                EtaDH = EtaDH,
                UWnd = UWnd,
                Region = Region,
                FValueC = FValueC, 
                FValueH = FValueH,
                y1 = y1,
                y2 = y2,
                **f) 
            for f in parts]
```

## 全体サンプリング

共通部分をまず生成させたあとに、その値を使ってほか部分を生成します。
最後の統合します。


```python
def sample():
    d_common = sample_common()
    d_walls = sample_wall(TotalFloorArea = d_common['TotalFloorArea'])
    d_windows = sample_window(TotalFloorArea = d_common['TotalFloorArea'], Region = d_common['Region'])
    d_doors = sample_door()
    d_perimeters = sample_earthfloor_perimeters()
    return { 
        'Common': d_common,
        'Walls' : d_walls,
        'Windows' : d_windows,
        'Doors' : d_doors,
        'EarthfloorPerimeters' : d_perimeters
    }
```
