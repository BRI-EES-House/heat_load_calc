## 2. 外皮性能計算プログラム＋簡単な追加情報

- 外皮性能計算プログラムの入力情報に、用途別床面積を追加したものより、推定を行う。
- 現状の外皮性能計算プログラムで入力されるのは用途別に分かれていない外皮性能であるため、室用途別の性能を求めるためには、外皮を用途別に割り振る必要がある。
- 住戸全体の各方位の外皮は、室用途別の床面積に応じて各用途に割り当てる。集合住宅の戸境壁等も同様の扱いとする。外皮性能は用途によらず均一に割り付ける。開口部は、2.外皮性能計算プログラムの入力情報より推定を開始する場合は、庇等の日射遮蔽部材を考慮した計算を行うため、方位、遮蔽部材やその寸法等が異なる場合は、別部材として扱う。その他は3.と同様に求める。
 
$$ \qquad
A_{主,i,j}= \frac{A_{i,j}・A_主}{A_主 + A_他 + A_非}, \quad
A_{他,i,j}= \frac{A_{i,j}・A_他}{A_主 + A_他 + A_非}\quad, \quad
A_{非,i,j}= \frac{A_{i,j}・A_主}{A_非 + A_他 + A_非}\quad
\qquad (1),\;(2),\;(3)
\\
$$
$$ \qquad
L_{主,i,j}= \frac{L_{i,j}・A_主}{A_主 + A_他 + A_非}, \quad
L_{他,i,j}= \frac{L_{i,j}・A_他}{A_主 + A_他 + A_非}\quad, \quad

- 入力値：

  - 主居室、その他の居室、非居室の床面積$[m ^ 2]$：$A_主$、$A_他$、$A_非$
  - 外皮の方位j、部位（一般部位又は開口部）𝑖の面積$[m ^ 2]$：$A_{i,j}$
  - 熱橋及び土間床等の外周部jの長さ$[m]$：$L_{i,j}$
  - 外皮の方位j、部位（一般部位又は開口部）iの熱貫流率$[W/m ^ 2 K]$：$U_{i,j}$
  - 熱橋及び土間床等の外周部jの線熱貫流率$[W/mK]$：$\psi _i$$_,$$_j$
  

- 出力値：

  - 主居室、その他の居室、非居室の外皮の方位j、部位（一般部位又は開口部）iの面積$[m ^ 2]$：$A_{主,i,j}$、$A_{他,i,j}$、$A_{非,i,j}$
  - 主居室、その他の居室、非居室の：熱橋及び土間床等の方位jの外周部iの長さ$[m]$：$L_{主,i,j}$、$L_{他,i,j}$、$L_{非,i,j}$
  - 主居室、その他の居室、非居室の外皮の方位j、部位（一般部位又は開口部）iの熱貫流率$[W/m ^ 2 K]$：$U_{主,i,j}$、$U_{他,i,j}$、$U_{非,i,j}$   
  - 主居室、その他の居室、非居室の熱橋及び土間床等の方位jの外周部iの線熱貫流率$[W/mK]$：$\psi _{主,i,j}$、$\psi _{他,i,j}$、$\psi_{非,i,j}$

```python
""" 室用途別の面積の算出 """
def Calc_Apu(Ap, Au, Aa):
    Apu =  Ap*Au/Aa
    return Apu

print (Calc_Apu(1, 1, 5))
```

```python
""" 室用途別の長さの算出 """
def Calc_Lpu(Lp, Au, Aa):
    Lpu =  Lp*Au/Aa
    return Lpu

print (Calc_Lpu(1, 1, 5))
```

```python
""" 室用途別の熱貫流率の算出 """
def Calc_Upu(Up):
    Upu =  Up
    return Upu

print (Calc_Upu(1))
```

```python
""" 室用途別の線熱貫流率の算出 """
def Calc_Ψpu(Ψp):
    Ψpu =  Ψp
    return Ψpu

print (Calc_Ψpu(1))
```

```python
""" 室用途別の内壁面積の算出のテスト1 """
h=2.8                  #階高[m]
a=[1.2,1.4,1.4]        #周長を求める際の係数[-]
Au=[29.8,51.3,38.9]    #主居室、その他の居室、非居室の床面積[㎡]
Aou=[78,126.50,114.50] #主居室、その他の居室、非居室の外皮の面積の合計[㎡]
Aiuumin=[0,0,0]        #室間の内壁等の面積の合計の最小値[㎡]
    
Aiu =  [max(0,2*Au[i]+4*h*a[i]*Au[i]**0.5-Aou[i]) for i in range(0,3)]

if Au[0]>0 and Au[2]>0 :
    Aiuumin[1]=2
if Au[1]>0 and Au[2]>0 :
    Aiuumin[2]=2
    
Aiuu=[0,0,0]
for i in range(0,3):
    Aiuu[i]=max(Aiuumin[i],(Aiu[0]+Aiu[1]+Aiu[2]-2*Aiu[-i-1])/2)
   #Aiuu[0]=max(Aiuumin[0],(Aiu[0]+Aiu[1]-Aiu[2])/2)
   #Aiuu[1]=max(Aiuumin[1],(Aiu[0]-Aiu[1]+Aiu[2])/2)
   #Aiuu[2]=max(Aiuumin[2],(-Aiu[0]+Aiu[1]+Aiu[2])/2)    
print (Aiuu)
```

```python
""" 室用途別の内壁面積の算出のテスト2 """
h=2.8                  #階高[m]
a=[1.2,1.4,1.4]        #周長を求める際の係数[-]
Au=[29.8,51.3,38.9]    #主居室、その他の居室、非居室の床面積[㎡]
Aou=[78,126.50,114.50] #主居室、その他の居室、非居室の外皮の面積の合計[㎡]
Aiuumin=[0,0,0]        #室間の内壁等の面積の合計の最小値[㎡]

def Calc_Aiuu(Au,Aou):
    for i in range(0,2):
        if Au[i]>0 and Au[2]>0:
            Aiuumin[i]=2    
    print(Aiuumin) 
    
    Aiu =  [max(0,2*Au[i]+4*h*a[i]*Au[i]**0.5-Aou[i]) for i in range(0,3)]
    Aiuu =  [max(Aiuumin[i],(Aiu[0]+Aiu[1]+Aiu[2]-2*Aiu[-i-1])/2) for i in range(0,3)]
    return Aiuu

print(Calc_Aiuu(Au,Aou))
```

### 換気・漏気

- 常時換気の経路は、居室より給気し、非居室より排気するものとする。風量は、住戸全体の気積（延床面積×2.4m）を基準として、その0.5回/hとする。各居室へ流入する風量は、住戸全体の風量を、各居室の床面積で案分した値とする。
- 漏気は、自立循環型住宅への設計ガイドライン（改修版）の検討が完了した時点で、それを参考に決定する。

$$ \qquad 
Q=(A_{主}+A_{他}+A_{非})×2.4×0.5
\qquad (**)
\\
$$
$$ \qquad 
Q_{主}=\frac{Q×A_{主}}{A_{主}+A_{他}}　\quad (**), \quad　Q_{他,i}=\frac{Q×A_{他,i}}{A_{主}+A_{他}}　\qquad (**)
\\
$$
$$ \qquad 
Q_{他}=\sum_{i}^{}Q_{他,i}　\quad (**), \quad　Q_{非}=Q_{主}+Q_{他}　\qquad (**)
\\
$$

- 入力値：

  - 主居室、その他の居室、非居室の床面積$[m ^ 2]$：$A_主$、$A_他$、$A_非$
  - その他の居室のうち、第i番目の居室の床面積$[m ^ 2]$：$A_{他,i}$

- 出力値：

  - 住戸全体の換気風量$[m ^ 3/h]$：$Q$
  - 主居室、その他の居室、非居室の換気風量$[m ^ 3/h]$：$Q_{主}$、$Q_{他}$、$Q_{非}$
  - その他の居室のうち、第i番目の居室の換気風量$[m ^ 3/h]$：$Q_{他,i}$

```python
""" 内壁の面積・熱貫流率の算出 """
Ai13i=[1,2,3,4,5]      #主たる居室―非居室間の内壁の第i部位の面積[m2](要素数は可変)
Ai23i=[1,2,3,4,5,6]    #その他の居室―非居室間の内壁の第i部位の面積[m2](要素数は可変)
Ar=[8,8,6,6]           #その他の居室のうち、和室、寝室、子供室1、子供室2の床面積割合[-]
Ui13i=[1,2,3,4,5]      #主たる居室―非居室間の内壁の第i部位の熱貫流率[W/(m2K)](要素数はAi13iに一致)
Ui23i=[1,2,3,4,5,6]      #主たる居室―非居室間の内壁の第i部位の熱貫流率[W/(m2K)](要素数はAi13iに一致)

def Calc_Aiuu(Aiuui):    #主たる居室―非居室間、その他の居室―非居室間の内壁面積の合計[m2]
    Aiuu =  sum(Aiuui)
    return Aiuu

def Calc_Uaiuu(Aiuui,Uiuui):    #主たる居室―非居室間、その他の居室―非居室間の内壁平均熱貫流率[W/(m2K)]
    try:
        Uaiuu = sum([Aiuui[i]*Uiuui[i] for i in range(0,len(Aiuui))]) / sum(Aiuui)
    except ZeroDivisionError as e:
        Uaiuu=0
    return Uaiuu

def Calc_Air23(Ai23):    #和室―非居室間、寝室―非居室間、子供室1―非居室間、子供室2―非居室間の内壁面積の合計[m2]
    Air23 =  [Ai23*i/sum(Ar) for i in Ar]
    return Air23

def Calc_Uari23(Uai23):    #和室―非居室間、寝室―非居室間、子供室1―非居室間、子供室2―非居室間の内壁平均熱貫流率[W/(m2K)]
    Uari23 = [Uai23 for i in Ar]
    return Uari23

print("Ai13={0}".format(Calc_Aiuu(Ai13i)))
print("Ai23={0}".format(Calc_Aiuu(Ai23i)))
print("Air23={0}".format(Calc_Air23(Calc_Ai23(Ai23i))))
print("Uai13={0}".format(Calc_Uaiuu(Ai13i,Ui13i)))
print("Uai23={0}".format(Calc_Uaiuu(Ai23i,Ui23i)))
print("Uari23={0}".format(Calc_Uari23(Calc_Uaiuu(Ai23i,Ui23i))))
```

```python
""" 換気量・漏気量の算出のテスト """
h=2.4                  #天井高[m]
n=0.5                  #換気回数[回/h]
Au=[29.8,51.3,38.9]    #主たる居室、その他の居室、非居室の床面積[㎡]
Ar=[8,8,6,6]           #その他の居室のうち、和室、寝室、子供室1、子供室2の床面積割合[-]

Au2=[Au[1]*i/sum(Ar) for i in Ar]  #その他の居室のうち、和室、寝室、子供室1、子供室2の床面積[㎡]

Q=sum(Au)*h*n                                   #住戸全体の換気風量[m3/h]
Q1=Q*Au[0]/(Au[0]+Au[1])                        #主たる居室の換気風量[m3/h]
Q2= [Q*Au2[i]/(Au[0]+Au[1]) for i in [0,1,2,3]] #その他の居室の換気風量[m3/h]
Q3=Q1+sum(Q2)                                   #非居室の換気風量[m3/h]

print ("Q={0}".format(Q))
print ("Q1={0}".format(Q1))
print ("Q2={0}".format(Q2))
print ("Q3={0}".format(Q3))
```
