```python
import sys
from matplotlib import pyplot as plt
import numpy as np

sys.path.append('../..')
from heat_load_calc import building
from heat_load_calc.building import Structure, Building
```

# UA値からC値を推定する

```python
# UA値の設定
ua_values = np.arange(1, 100) / 100.0
```

```python
# RC造のC値
rc_c_values = Building._estimate_c_value(ua_value=ua_values, struct=Structure.RC)

# SRC造のC値
src_c_values = Building._estimate_c_value(ua_value=ua_values, struct=Structure.SRC)

# 木造のC値
wooden_c_values = Building._estimate_c_value(ua_value=ua_values, struct=Structure.WOODEN)

# S造のC値
s_c_values = Building._estimate_c_value(ua_value=ua_values, struct=Structure.STEEL)
```

```python
# グラフの描画
fig, ax = plt.subplots(2, 2, figsize=(12, 8))
ax[0,0].set_xlabel('UA value[W/m2K]')
ax[0,1].set_xlabel('UA value[W/m2K]')
ax[1,0].set_xlabel('UA value[W/m2K]')
ax[1,1].set_xlabel('UA value[W/m2K]')

ax[0,0].set_ylabel('C value[cm2/m2]')
ax[0,1].set_ylabel('C value[cm2/m2]')
ax[1,0].set_ylabel('C value[cm2/m2]')
ax[1,1].set_ylabel('C value[cm2/m2]')

ax[0,0].set_title('RC')
ax[0,1].set_title('SRC')
ax[1,0].set_title('Wooden')
ax[1,1].set_title('Steel')

ax[0,0].plot(ua_values, rc_c_values)
ax[0,1].plot(ua_values, src_c_values)
ax[1,0].plot(ua_values, wooden_c_values)
ax[1,1].plot(ua_values, s_c_values)

plt.tight_layout()
plt.show()
```
