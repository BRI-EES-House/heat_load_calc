```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_tau_satoh(phi: float) -> float:
    """
    Args:
        phi: 入射角, degree
    Returns:
        入射角に対する基準化透過率
    """
    cos_phi = np.cos(np.radians(phi))
    return (2.392 * cos_phi - 3.8636 * cos_phi ** 3 + 3.7568 * cos_phi ** 5 - 1.3965 * cos_phi ** 7) / 0.88
```

```python
def get_tau_jis(phi: float) -> float:
    """
    Args:
        phi: 入射角, degree
    Returns:
        入射角に対する基準化透過率
    """
    cos_phi = np.cos(np.radians(phi))
    return 0.000 * cos_phi ** 0 + 2.552 * cos_phi ** 1 + 1.364 * cos_phi ** 2 - 11.388 * cos_phi ** 3 + 13.617 * cos_phi ** 4 - 5.146 * cos_phi ** 5
```

```python
def get_rho_jis(phi: float) -> float:
    """
    Args:
        phi: 入射角, degree
    Returns:
        入射角に対する基準化反射率
    """
    cos_phi = np.cos(np.radians(phi))
    return 1.000 * cos_phi ** 0 - 5.189 * cos_phi ** 1 + 12.392 * cos_phi ** 2 - 16.593 * cos_phi ** 3 + 11.851 * cos_phi ** 4 - 3.461 * cos_phi ** 5    
```

```python
def get_tau_jis_double(phi: float) -> float:
    """
    Args:
        phi: 入射角, degree
    Returns:
        入射角に対する基準化透過率
    """
    tau = get_tau_jis(phi)
    rho = get_rho_jis(phi)
    return tau * tau / (1 - rho * rho)
```

```python
plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = fig.subplots()

phi = np.arange(start= 0.0, stop= 91.0, step = 1.0)
ax.plot(phi, get_tau_satoh(phi), label='Satoh')
ax.plot(phi, get_tau_jis(phi), label='JIS transmittant')
ax.plot(phi, get_rho_jis(phi), label='JIS refrect')
ax.plot(phi, get_tau_jis_double(phi), label='JIS transmittant double')

ax.set_xlabel('angle, degree')
ax.set_ylabel('standardized coefficient')

plt.legend()
plt.show()
```
